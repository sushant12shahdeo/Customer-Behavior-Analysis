```python
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda  x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns=df.columns.str.lower()
```

```python
df.columns=df.columns.str.replace(" ","_")
```

```python
df=df.rename(columns={"purchase_amount_(usd)": "Purchase_Amount"})
```

```python
df.columns
```

```python
#creating a age-group column
labels=["Young-Adults","Adults","Middle-age","Senior"]
df["age-group"]=pd.qcut(df["age"], q=4, labels=labels)
```

```python
df[["age","age-group"]].head()
```

```python
#creating a column frequency of days
frequency_mapping={
    "Fortnightly":14,
    "Weekly":7,
    "Monthly":30,
    "Quarterly":90,
    "Bi-Weekly":14,
    "Annually":365,
    "Every 3 Months":90
}
df["frequency_of_days"]=df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[["frequency_of_days","frequency_of_purchases"]].head(10)
```
